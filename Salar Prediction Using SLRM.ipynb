```python
df[['Salary', 'YearsExperience']].corr()
```

```python
X = df[['YearsExperience']]
Y = df[['Salary']]
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X,Y)
```

```python
y_pred = model.predict(X)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
MSE = mean_squared_error(Y, y_pred)
print("Mean Square Error :", np.round(MSE,2))
print("Root Mean Square Error :", np.round(np.sqrt(MSE),2))
R2 = r2_score(Y, y_pred)
print("R Square value :", np.round(R2, 2))
print("Model Accuracy :", np.round(R2,2)*100,"%")
```

```python
plt.scatter(df['YearsExperience'], df['Salary'], color = 'orange')
plt.scatter(df['YearsExperience'], y_pred, color='blue')
plt.plot(df['YearsExperience'], y_pred, 'k--', color = 'red')
plt.xlabel("Years Experience")
plt.ylabel("Predicted Salary")
plt.title("Actual vs Predicted Salary")
plt.tight_layout()
plt.show()
```

The Data points in yellow color represents the "Actual Values" of the Salary column (X)


The Data points in Blue color represents the "Predicted Values" of the Salary column (X)

```python
#B0 value
B0 = np.round(model.intercept_,2)
B0
```

```python
#B1 value
B1 = np.round(model.coef_,2)
B1
```
